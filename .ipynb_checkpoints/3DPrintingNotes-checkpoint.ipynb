## Day 2, (6.23.2016):
### SLA, Micro SLA
- Vat of liquid UV-curable resin
- DLP light exposure leads to material hardening
- *Requires* support structures
- Good for jewlery and investment cast

### Selective Laser Sintering (SLS)
- Wiping material across surface $\rightarrow$ melt material
- Rising part, lots of cleaning
- Polymer powder fused together
- Already has support structure
- Always have thin walls and a drain
- Chamber is heated and inert
- Curing of photopolymers
- Powder can be re-used (w/ a certain amount of used and un-used)
- Close to injection molded material properties

### Directed Metal Printing (DMP, DMLS)
- Need support
- Metal powder $\rightarrow$ electron beam w/ heated or intert gas
- Powdered aluminum can react poorly and explode
- Waste material is a big deal

### MultiJet Printing (MJP)
- *Requires* support structure
- Layer of polymer
- Also Polyjet from Stratasys

### (CJP)
- Fine powder, plaster or gypsum
- Fast
- Already has support
- Really fragile

- **Bending:** As it sounds
- **Deep Drawing:** Plastic forming of materials; Pressing

### Plastics:

- **Polymers:** Large molecules, no vapor phase, strong intermolecular forces
- Crystalline or amorphous
- **Highly Crystalline:** rigid, high melting
- **Amorphous:** No long range order
- **Glass Transition:** *Reversible* transition of a material from hard to rubber-like
- **Slope:** Solid specific heat vs. liquid specific heat
- **DMA:** For measuring loss of energy through 'bounce'
- Thermoplastic vs. thermosetting
- Calculus *vs.* Monte Carlo *vs.* Measured Signal to Noise

### CubePro
- Build $\rightarrow$ replace
- Sensitive to bed level
- Move bed up $\rightarrow$ home flag switch (back right) to home z axis $\rightarrow$ apply glue $\rightarrow$ nozzle to center $\rightarrow$ heat up $\rightarrow$ over wiper $\rightarrow$ purge then wipe

- Fan's needed for PLA
- ABS doesn't like to be cooled off
- Cooling algorythm depending on raft presence
